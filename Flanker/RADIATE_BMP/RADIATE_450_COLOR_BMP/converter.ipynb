```python
# --- Notebook cell 1: Install deps (run once) ---
# Pillow is the standard image library. If you already have it, this is harmless.
%pip -q install pillow

# --- Notebook cell 2: Imports & config ---
from pathlib import Path
from PIL import Image

# Config
ROOT = Path.cwd()              # current directory
RECURSIVE = True               # include all child folders
OVERWRITE = False              # if True, overwrite existing .webp
DELETE_ORIGINAL = False        # if True, delete .bmp after successful conversion

# WebP options
QUALITY = 90                   # 0-100 (higher = better quality, larger file)
LOSSLESS = False               # True for lossless WebP (often larger)
METHOD = 6                     # 0-6 (higher = better compression, slower)

# --- Notebook cell 3: Discover BMPs ---
bmp_paths = sorted(ROOT.rglob("*.bmp")) if RECURSIVE else sorted(ROOT.glob("*.bmp"))

print(f"Root: {ROOT}")
print(f"Found {len(bmp_paths)} .bmp file(s).")

# Preview a few
for p in bmp_paths[:10]:
    print(" -", p.relative_to(ROOT))
if len(bmp_paths) > 10:
    print(" ...")

# --- Notebook cell 4: Convert ---
converted = 0
skipped = 0
failed = 0
failures = []

for bmp_path in bmp_paths:
    webp_path = bmp_path.with_suffix(".webp")

    if webp_path.exists() and not OVERWRITE:
        skipped += 1
        continue

    try:
        with Image.open(bmp_path) as im:
            # Ensure a WebP-friendly mode; preserve alpha if present
            if im.mode not in ("RGB", "RGBA"):
                # If the image has transparency, convert to RGBA; otherwise RGB
                im = im.convert("RGBA" if "A" in im.getbands() else "RGB")

            im.save(
                webp_path,
                format="WEBP",
                quality=QUALITY,
                lossless=LOSSLESS,
                method=METHOD,
            )

        converted += 1

        if DELETE_ORIGINAL:
            bmp_path.unlink()

    except Exception as e:
        failed += 1
        failures.append((bmp_path, repr(e)))

print(f"Converted: {converted}")
print(f"Skipped (already existed): {skipped}")
print(f"Failed: {failed}")

if failures:
    print("\nFailures (first 10):")
    for p, err in failures[:10]:
        print(" -", p.relative_to(ROOT), "->", err)

# --- Notebook cell 5 (optional): Sanity check counts ---
webp_paths = sorted(ROOT.rglob("*.webp")) if RECURSIVE else sorted(ROOT.glob("*.webp"))
print(f"Now found {len(webp_paths)} .webp file(s) under {ROOT}.")
```
